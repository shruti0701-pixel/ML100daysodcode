### Categorical ScatterPlots ###

```python
# x-axis ---> Categorical data
# y-axis ---> Numerical Data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips =sns.load_dataset('tips')  
```

```python
tips.head()
```

## Strip Plot ##

```python
sns.catplot(x='day',y='tip',kind='strip',data=tips)


#jitter(algorithm) - induces small noise between two points 
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=2 ,data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=10 ,data=tips)
```

```python
sns.catplot(x='day',y='tip',kind='strip',jitter=0 ,data=tips)
```

## Swarm Plot ##

```python
sns.catplot(x='day',y='tip',kind='swarm' ,data=tips)
```

```python
# wheneever it finds out that there are multiple values occupying the same position, it represnets the distribution of data 
# symmetric distribution on both direction
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue ='sex',data=tips)
```

```python
## no style and size
```

```python
sns.swarmplot(x='day',y='tip',data=tips) ##rather than using catplot directly you can use swarmplot
```
