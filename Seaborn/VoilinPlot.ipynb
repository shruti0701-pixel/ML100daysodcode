## Violin Plots ##

violin plots are amethod of plotting numerical data and can be considered a combination of the box plot with a kernel density plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight') #styling element
```

```python
tips =sns.load_dataset('tips')  
```

```python
tips.head()
```

## violin plots are often used to compare the distribution of a given variable across some categories ##

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue ='sex',kind='violin',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',hue ='sex',kind='violin',split=True,data=tips)
```
