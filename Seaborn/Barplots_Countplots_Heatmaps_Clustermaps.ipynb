```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips =sns.load_dataset('tips')  
```

```python
tips.head()
```

```python
sns.catplot(x='smoker' , y='tip',kind='bar' , data = tips)
```

```python
sns.catplot(x='smoker' , y='total_bill',kind='bar' , data = tips)
```

```python
sns.catplot(x='smoker' , y='total_bill',hue = 'sex',kind='bar' , data = tips)
```

```python
sns.catplot(x='smoker' , y='total_bill',hue = 'sex',kind='bar',estimator=np.median , data = tips)
```

```python
sns.catplot(x='smoker' , y='total_bill',hue = 'sex',kind='bar',estimator=np.var , data = tips)
```

```python
sns.catplot(x='smoker' , y='total_bill',hue = 'sex',kind='bar',estimator=np.std , data = tips)
```

## CountPlot ##
- special case of bar plot
- where you do not consider numerical value, you only consider categorical value and you find to the count of the value like histogram.

```python
sns.catplot(x='sex',kind='count',data=tips)
```

```python
sns.catplot(x='sex',hue = 'smoker',kind='count',data=tips)
```

## HeatMaps ##
- A heat map is a graphical representation of data where the individual values contained i a matrix are represented as colors.
- two categorical data , sometimes numerical data but it should be discrete.

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x , cbar=False)
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5)
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 , annot=True , fmt='d')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 , annot=True , fmt='d' , cmap = 'viridis')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 , annot=True , fmt='d' , cmap = 'summer')
```

```python
sns.heatmap(x , cbar=False , linewidths=0.5 , annot=True , fmt='d' , cmap = 'winter')
```

## Clustermaps ##
- similar rows , colums will come together

```python
sns.clustermap(x)
```

```python
sns.clustermap(x , metric = 'correlation')
```

```python
sns.clustermap(x , z_score = 0, annot = True, metric = 'correlation')
```
