## Jointplot ##
Jointplot creates a multi-panel figure that shows both the bivariate(or joint) relationship between two variables along with the univariate(or marginal) distribution of each on separate axes.

We can use Jointplots in the following manners:
1. Scatter Plots
2. Hexbin Plots
3. KdePlots
4. RegPlot - Regression Plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , data=tips)  #individual histograms
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , kind = 'hex' ,data=tips)   ##hexagonal plot - 2D histogram
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , kind = 'kde' ,data=tips)   ##contourplots - 2d PDF (Probablity Density Function)
```

```python
## inner  - higher points on the peak
## outer  - lower points on the peak
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , kind = 'reg' ,data=tips)   
```

```python
## if you give me tip I can predict total bill - BEST FIT LINE
```

## Pairplots ##
- A pair plot is a visual summary of all the relationships between pairs of variables in your data.
- It's like having a bunch of two-way conversations between your data points, all at once!

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris , hue ='species')
```

## DistPlots ##
- used to do univariate analysis on a continuous numerical data
- Distplots combines 3 things in one graph:
  1. Histogram
  2. Kernel Density Estimation(PDF)
  3. Rugplot

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

## Plotting a Distplot ##

```python
sns.distplot(titanic['age'])
```

## Plotting only the Histogram ##

```python
sns.distplot(titanic['age'],kde=False)
```

```python
sns.distplot(titanic['age'],bins =10,kde=False)
```

```python
sns.distplot(titanic['age'],bins =100,kde=False)
```

## Plotting only the KDE ##

```python
sns.distplot(titanic['age'],hist=False)
```

## Plotting only the Rugplot ##

```python
sns.distplot(titanic['age'],hist = False,kde=False , rug = True)  ##plotting small vertical line on every observation
```

## How to use DistPlot ##

```python
sns.distplot(titanic [titanic['survived'] == 1] ['age'])
```
