```python
## A boxplot is a way to quickly visualize how a bunch of numbers are spread out.
## It shows the middle, the spread around the middle, and any outliers.  
## Think of it like a box with whiskers showing where most of the data falls.
```

```python
## distirbution plot and should be plotted on top of a numerical data
## comparison between 2 categorical variables 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight') #styling element
```

```python
tips =sns.load_dataset('tips')  
```

```python
tips.head()
```

```python
#Univariate

sns.boxplot(tips['tip'])  ## blue part - interquatile range
```

```python
sns.boxplot(tips['total_bill'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```
